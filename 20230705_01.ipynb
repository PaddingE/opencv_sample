```python
import sys
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
cap = cv2.VideoCapture(0)

if not cap.isOpened():
    print('video capture failed')
    sys.exit()
    
w = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
h = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = int(cap.get(cv2.CAP_PROP_FPS))
fourcc = cv2.VideoWriter_fourcc(*'DIVX')

out = cv2.VideoWriter('review.avi', fourcc, fps,(w,h))
    
while True:
    ret, frame = cap.read()
    
    if not ret:
        print('frame read failed')
        break
    
    edge = cv2.Canny(frame, 50,150)
    edge_color = cv2.cvtColor(edge, cv2.COLOR_GRAY2BGR)
    
    out.write(edge_color)
    
    cv2.imshow('frame', frame)
    cv2.imshow('edge', edge)
    
    if cv2.waitKey(20) == 27:
        break
    
cv2.destroyAllWindows()
cap.release()
out.release()
```

```python
src = cv2.imread('./fig_ch2/cow.png')
dst = cv2.imread('./fig_ch2/green.png')

src_GRAY = cv2.cvtColor(src, cv2.COLOR_BGR2GRAY)
h,w = src.shape[:2]
dst_crop = dst[300:300+h, 300: 300+w]

ret, mask = cv2.threshold(src_GRAY, 240,255,cv2.THRESH_BINARY_INV)
cv2.copyTo(src, mask, dst_crop)

cv2.namedWindow('mask', cv2.WINDOW_NORMAL)
cv2.imshow('src', src)
cv2.imshow('dst', dst)
cv2.imshow('mask', mask)

cv2.waitKey()
cv2.destroyAllWindows()
```

```python
img = cv2.imread('./fig_ch2/puppy.bmp', cv2.IMREAD_GRAYSCALE)
img_base = img.copy()

cv2.namedWindow('img', cv2.WINDOW_AUTOSIZE)
cv2.imshow('img', img)

while True:
    key = cv2.waitKey()
    if key == ord('q') or key == 27:
        break
    
    elif key == ord('i'):
        img = 255 - img
        cv2.imshow('img', img)
        
    elif key == ord('c'):
        img = cv2.Canny(img, 50,100)
        cv2.imshow('img', img)
    
    elif key == ord('r'):
        img = img_base
        cv2.imshow('img',img)
    
    elif key == ord('s'):
        cv2.imwrite('./save_img.bmp', img)
    
cv2.destroyAllWindows()
```

```python
cap = cv2.VideoCapture(0)

if not cap.isOpened():
    print('video capture failed')
    sys.exit()

ret, frame = cap.read()
frame_base = frame.copy()

edge_checker = False
flip_checker = False
inverse_checker = False

while True:
    if edge_checker == True:
        ret, frame = cap.read()
        edge = cv2.Canny(frame, 50,150)
        edge_color = cv2.cvtColor(edge, cv2.COLOR_GRAY2BGR)
        frame = edge_color
        
    else:
        ret, frame = cap.read()
    
    if not ret:
        print('frame read failed')
        break
    
    if flip_checker == True:
        frame = cv2.flip(frame,1)
        
    if inverse_checker == True:
        frame = 255 - frame
    
    key = cv2.waitKey(20)
    
    if key == 27 or key == ord('q'):
        break
    
    elif key == ord('e'):
        edge_checker = not edge_checker
    
    elif key == ord('f'):
        flip_checker = not flip_checker
        
    elif key == ord('i'):
        inverse_checker = not inverse_checker
    
    cv2.imshow('frame', frame)
    
cv2.destroyAllWindows()
cap.release()
```
