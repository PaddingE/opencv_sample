```python
import cv2
import numpy as np
import sys
print(cv2.__version__)
```

```python
def show_img(img):
    ## 이미지가 없을때 발생할 오류를 예외처리
    if img is None:
        print('image read failed')
        sys.exit()

    ## 윈도우 이름 지어주고 이미지를 윈도우에 띄우기(flag의 default값 : cv2.WINDOW_AUTOSIZE)    
    cv2.namedWindow('img', cv2.WINDOW_AUTOSIZE)
    cv2.imshow('img', img)

    ## 키값 받기
    cv2.waitKey()

    ## 윈도우 닫기
    cv2.destroyAllWindows()
```

```python
## 이미지 불러오기(flag의 default값 : cv2.IMREAD_COLOR)
img = cv2.imread('./fig/puppy.bmp', cv2.IMREAD_GRAYSCALE)

## 이미지 크기 조절
img_resize = cv2.resize(img, (320,240), interpolation= cv2.INTER_AREA)

## 이미지 저장
cv2.imwrite('./fig/my_puppy_resize.png', img_resize)
show_img(img)
```

```python
print('type = ', type(img))
print('img shape = ', img.shape)
## unsigned int 8 : 0~255의 값으로 표현
print('dtype = ', img.dtype)
```

```python
img_BGR = cv2.imread('./fig/puppy.bmp', cv2.IMREAD_COLOR)
img_RGB = cv2.cvtColor(img_BGR, cv2.COLOR_BGR2RGB)
img_GRAY = cv2.cvtColor(img_BGR, cv2.COLOR_BGR2GRAY)

fig, ax = plt.subplots(1,3, figsize = (12,5))

ax[0].imshow(img_BGR), ax[0].axis('off'), ax[0].set_title('BGR')
ax[1].imshow(img_RGB), ax[1].axis('off'), ax[1].set_title('RGB')
ax[2].imshow(img_GRAY, cmap = 'gray'), ax[2].axis('off'), ax[2].set_title('GRAY')
plt.show()
```

```python
## 슬라이드 쇼

import os
```

```python
img_files = os.listdir('./fig/images/')

#print(img_files)

file_paths = []

for i in img_files:
    file_path = './fig/images/' + i
    file_paths.append(file_path)
    
#display(file_paths)

cv2.namedWindow('scene', cv2.WINDOW_NORMAL)
cv2.setWindowProperty('scene', cv2.WND_PROP_FULLSCREEN, cv2.WINDOW_FULLSCREEN)

idx = 0
while True:
    
    img = cv2.imread(file_paths[idx], cv2.IMREAD_COLOR)
    
    if img is None:
        continue
    
    cv2.imshow('scene', img)
        
    if cv2.waitKey(3000) == 27:
        break
    
    idx += 1
    
    if idx >= len(file_paths):
        idx = 0        
    
cv2.destroyAllWindows()
```
